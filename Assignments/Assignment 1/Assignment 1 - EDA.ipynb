#### Fill missing values of train_data_backup and print summary statistics.

```python
for col in train_data_backup.iloc[:, 4:].columns:
    mod = train_data_backup[col].mode()[0]
    train_data_backup[col].fillna(mod, inplace=True)
    print("Attribute:", col)
    print("Summary statistics")
    print(train_data_backup[col].describe())
    print("----------------------------")
```

#### Comment on which method is better and why.

The mode filling is a better approach as it allows healthy fluctutations in data which backward filling doesn't. The most frequent temperature changes are a better estimate compared to the coming temperature changes.
